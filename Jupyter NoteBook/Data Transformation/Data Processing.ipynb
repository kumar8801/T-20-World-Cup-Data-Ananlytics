**Match Results**

```python
import pandas as pd
import json
```

```python
with open('json files_data/t20_wc_match_results.json') as fj:
    data=json.load(fj)
data[0]["matchSummary"]
```

```python
df_match=pd.DataFrame(data[0]["matchSummary"])
df_match
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
match_ids_dict={}
for index,row in df_match.iterrows():
    key1=row['team1']+' Vs '+row['team2']
    key2=row['team2']+' Vs '+row['team1']
    
    match_ids_dict[key1]=row['match_id']
    match_ids_dict[key2]=row['match_id']
```

```python
df_match.to_csv('crickfiles/match_results.csv',index=False)
```

**Batting Summary**

```python
with open("json files_data/t20_wc_batting_summary.json") as f:
    data=json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec['battingSummary'])
df_batting=pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
```

```python
df_batting['match_id']=df_batting['match'].map(match_ids_dict)
df_batting.head(11)
```

```python
df_batting.to_csv('crickfiles/batting.csv',index=False)
```

**Bowling Summary**

```python
with open('json files_data/t20_wc_bowling_summary.json') as f:
    data=json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling=pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('crickfiles/bowling_summary.csv', index = False)
```

**Players Information**

```python
with open('json files_data/t20_wc_player_info.json') as f:
    data=json.load(f)
```

```python
df_players = pd.DataFrame(data)
print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv('crickfiles/players_info.csv',index=False)
```
